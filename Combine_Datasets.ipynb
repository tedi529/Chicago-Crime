```python
# Import Dependencies
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import os
```

## Import and Clean Chicago Crime Datasets

```python
# Import and clean 2016 crime data
crime_2016 = os.path.join("Resource", "crime_clean_2016.csv") 

crime_2016_df = pd.read_csv(crime_2016, index_col='id')

crime_2016_df[['year', 'month', 'daytime']] = crime_2016_df['date'].str.split("-", expand=True)
crime_2016_df[['day', 'time']] = crime_2016_df['daytime'].str.split("T", expand=True)
crime_2016_df['month_day'] = crime_2016_df["month"] + crime_2016_df["day"]
crime_2016_df_final = crime_2016_df.drop(columns=["location", "date", "month", "year", "daytime", "day", "time"], axis=1)
crime_2016_df_final
```

```python
# Import and clean 2017 crime data
crime_2017 = os.path.join("Resource", "crime_clean_2017.csv") 

crime_2017_df = pd.read_csv(crime_2017, index_col='id')

crime_2017_df[['year', 'month', 'daytime']] = crime_2017_df['date'].str.split("-", expand=True)
crime_2017_df[['day', 'time']] = crime_2017_df['daytime'].str.split("T", expand=True)
crime_2017_df['month_day'] = crime_2017_df["month"] + crime_2017_df["day"]
crime_2017_df_final = crime_2017_df.drop(columns=["location", "date", "month", "year", "daytime", "day", "time"], axis=1)
crime_2017_df_final
```

```python
# Import and clean 2018 crime data
crime_2018 = os.path.join("Resource", "crime_clean_2018.csv") 

crime_2018_df = pd.read_csv(crime_2018, index_col='id')

crime_2018_df[['year', 'month', 'daytime']] = crime_2018_df['date'].str.split("-", expand=True)
crime_2018_df[['day', 'time']] = crime_2018_df['daytime'].str.split("T", expand=True)
crime_2018_df['month_day'] = crime_2018_df["month"] + crime_2018_df["day"]
crime_2018_df_final = crime_2018_df.drop(columns=["location", "date", "month", "year", "daytime", "day", "time"], axis=1)
crime_2018_df_final
```

```python
# Check if columns in dataframes are identical 
if crime_2016_df_final.columns.all() == crime_2017_df_final.columns.all() == crime_2018_df_final.columns.all():
    print("Columns are identical")
else:
    print("Columns are not identical")
```

```python
# Join datasets for 2016, 2017, and 2018
join1 = crime_2016_df_final.append(crime_2017_df_final)
training_df = join1.append(crime_2018_df_final)
training_df
```

```python
# Check sum of lengths of individual datasets match combined length
length1 = len(crime_2016_df_final) + len(crime_2017_df_final) + len(crime_2018_df_final)
if length1 == len(training_df):
    print("Same length")
else: 
    print("Different length")
```
